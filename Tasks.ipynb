```python
import pandas as pd
import numpy as np
```

```python
data = {
    "ID" : [101,102,103,104,105],
    "Name" : ["Sara","Ahmed","John","Maria","Ali"],
    "Role" : ["AI Engineer","Data Analyst","ML Ops","AI Engineer","Data Analyst"],
    "Salary" : [95000,72000,88000,105000,65000],
    "Experience (Yrs)": [3,1,5,4,2]
}
df = pd.DataFrame(data)
df_filtered = df[(df["Role"] == "AI Engineer") & (df["Experience (Yrs)"]> 3)]
print(df_filtered)
print(df.groupby("Role")["Salary"].mean())
for x in df.index:
    if df.loc[x,"Experience (Yrs)"] >= 4:
        df.loc[x,"Level"]="Senior"
    else:
        df.loc[x,"Level"]="Junior"
print(df)
```

```python
arr = np.random.randint(10,101,(5,5))
print("Original Array")
print(arr)
print("values greater than avg")
for i in np.nditer(arr):
    if i > arr.mean():
        print(i)
new_arr = arr
new_arr[new_arr < 50 ] = 0
print(new_arr)
```

```python
df.loc[1,"Salary"] = None
print(df)
x = df["Salary"].median()
df["Salary"] = df["Salary"].fillna(x)
print(df)
df = df.sort_values(by="Salary",ascending = False).reset_index(drop = True)
print(df)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
df_sel = df[["Salary", "Experience (Yrs)"]]
corr = df_sel.corr(numeric_only = True)
sns.heatmap(corr,annot=True)
plt.show()
```

```python
sns.scatterplot(x="Experience (Yrs)", y="Salary", hue = "Role", data = df)
plt.title("Salary vs Experience in AI roles")
plt.grid(True)
plt.show()
```

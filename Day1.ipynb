```python
marks = {
    "Yusra": [22, 88, 95],
    "Hafsa": [70, 75, 80],
    "Maria": [98, 82, 91],
}
avg = {}
for key in marks.keys():
    avg[key] = 0
for key,value in marks.items():
    sum = 0
    for i in value:
        sum = sum + i
    num = sum/300
    avg[key] = num
print(avg)
print("Top Performer: ")
print(max(avg, key= avg.get))
```

```python
library = {
    "Fiction": {"books": ["Dune", "1984", "Brave New World"], "members": ["Ali", "Yusra", "Ali", "Hafsa"]},
    "Science": {"books": ["Cosmos", "Brief History", "Cosmos"], "members": ["Maria", "Sadia", "Maria"]},
    "History": {"books": ["Sapiens", "Guns Germs Steel"], "members": ["Ali", "Hafsa", "Sadia"]}
}
print("Total number of books: ")
count = 0
for genera, value in library.items():
    for i in value["books"]:
        count = count +1;
print(count)
print(library["Fiction"]["books"][0])
dup = ()
print("genera with duplicates: ")
for genera, value in library.items():
    has_dup = False
    for i in range(0,len(value["books"])):
        for j in range(i+1,len(value["books"])):
            if value["books"][i] == value["books"][j] :
                dup = list(dup)
                dup.append(value["books"][i])
                dup = tuple(dup)
                has_dup = True
    if (has_dup):
        print(genera)
print("duplicates: ")
print(dup)
set1 = {}
for genera, value in library.items():
    set1.clear()
    print(genera)
    set1 = set(value["members"])
    print("Members: ")
    print(set1)
All_members = []
for genera,value in library.items():
    for i in value["members"]:
        All_members.append(i)
print(All_members)
my_dict = {}
done = []
for i in range(0,len(All_members)):
    count = 1
    if(All_members[i] in done):
        continue
    for j in range(i+1,len(All_members)):
        if All_members[i] == All_members[j] :
            count = count+1
    my_dict[All_members[i]] = count
print("Appears most: ")
print(max(my_dict,key = my_dict.get))            

gen = []
for genera, value in library.items():
    set2 = set(value["members"])
    if len(set2)> 2:
        gen.append(genera)
gen = tuple(gen)
print(gen)
```

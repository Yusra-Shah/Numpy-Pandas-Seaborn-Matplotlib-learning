```python
#Matplotlib
import matplotlib.pyplot as plt
```

```python
import numpy as np
x = np.array([1,2,3,4,5])
y = np.array([10,20,30,40,50])
plt.plot(x,y)
plt.show()
```

```python
x = np.array([1,2,3,4,5])
y = np.array([1,2,3,4,5])
plt.plot(x,y)
plt.show()
```

```python
x = np.array([-3,-2,-1,0,1,2,3])
y = np.array([9,4,1,0,1,4,9])
plt.plot(x,y)
plt.show()
```

```python
x=[1,2,3]
y=[2,4,6]
z=[1,3,5]
plt.plot(x,y,z)
plt.show()
```

```python
x = np.array([1,2,3,4,5])
y = np.array([1,2,3,4,5])
plt.plot(x,y)
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Linear")
plt.show()
```

```python
x = np.array([1,2,3,4,5])
y = np.array([1,2,3,4,5])
plt.plot(x,y,linestyle='--',marker = 'o')
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Customized Linear")
plt.show()
```

```python
x = np.array([1,2,3,4,5])
y = np.array([1,2,3,4,5])
y2 = np.array([1,4,9,16,25])
plt.plot(x,y,label="line 1")
plt.plot(x,y2, label="line 2")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Linear")
plt.legend() #to show legends
plt.show()
```

```python
#Bar charts
categories = ["Apple","Mango","Banana"]
amount = [24,67,34]
plt.bar(categories,amount)
plt.title("Bar chart")
plt.show()
```

```python
data = np.random.randn(100)
plt.hist(data,bins=20)
plt.title("Histogram")
plt.show()
```

```python
x = np.random.rand(25)
y = np.random.rand(25)
plt.scatter(x,y)
plt.show()
```

```python
x = np.random.rand(25)
y = np.random.rand(25)
plt.figure(figsize =(10,8) )
plt.plot(x,y)
plt.show()
```
